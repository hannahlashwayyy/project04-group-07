```python
# Import dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

```python
# Read the data into a Pandas DataFrame
diabetes_df = pd.read_csv('diabetes.csv') 
```

```python
diabetes_df.head()
```

```python
#combining prediabetic and diabetic values 
diabetes_df['Diabetes_012'] = diabetes_df['Diabetes_012'].replace(1, 2)

# Converting diabetic value to 1 
diabetes_df['Diabetes_012'] = diabetes_df['Diabetes_012'].replace(2, 1)

#double checking we only have 2 values now in the diabetes column
diabetes_df['Diabetes_012'].value_counts()
```

```python
#fixing education column 
diabetes_df['Education'] = diabetes_df['Education'].replace({1: 3, 2: 3})

#readjusting column values
diabetes_df['Education'] = diabetes_df['Education'].replace({3: 1, 4: 2, 5: 3, 6: 4 })

#checking values 
diabetes_df['Education'].value_counts()
```

```python
diabetes_df.info()
```

```python
# Fill NaN values with a specific integer (e.g., 0) before converting
diabetes_df = diabetes_df.fillna(0)

# Convert relevant float columns to integer type
float_columns = [
    'Diabetes_012', 'HighBP', 'HighChol', 'CholCheck', 'Smoker',
    'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Fruits',
    'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost',
    'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk', 'Sex', 'Age',
    'Education', 'Income'
]

# Convert to int
diabetes_df[float_columns] = diabetes_df[float_columns].astype(int)

# Check the updated DataFrame
print(diabetes_df.dtypes)
print(diabetes_df.head())
```

```python
# Dictionaries to convert column values from binary to strings
# Create a mapping for each column

# Diabetes column
diabetes_map = {0: 'not diabetic', 1: 'diabetic'}

# HighBP column
bp_map = {0: 'no', 1: 'yes'}

# HighChol column
chol_map = {0: 'no', 1: 'yes'}

# CholCheck column
cholcheck_map = {0: 'no', 1: 'yes'}

# Smoker column
smoke_map = {0: 'no', 1: 'yes'}

# Stroke column
stroke_map = {0: 'no', 1: 'yes'}

# HeartDiseaseorAttack column
heart_map = {0: 'no', 1: 'yes'}

# PhysActivity column
activity_map = {0: 'no', 1: 'yes'}

# Fruits column
fruit_map = {0: 'no', 1: 'yes'}

# Veggies column
veggies_map = {0: 'no', 1: 'yes'}

# HvyAlcoholConsump column
alcohol_map = {0: 'no', 1: 'yes'}

# AnyHealthcare column
healthcare_map = {0: 'none', 1: 'has healthcare'}

# NoDocbcCost column
dr_map = {0: 'no', 1: 'yes'}

# General health column
health_map = {1: 'excellent', 2: 'very good', 3: 'good', 4: 'fair', 5: 'poor'}

# Difficulty walking column
walking_map = {0: 'no difficulty', 1: 'difficult'}

# Sex column
sex_map = {0: 'female', 1: 'male'}

# Age column
age_map = {1: '18-24', 2: '24-29', 3: '30-34', 4: '35-39', 5: '40-44', 6: '45-49',
           7: '50-54', 8: '55-59', 9: '60-64', 10: '65-69', 11: '70-74', 12: '75-79',
           13: '80+'}

# Education column
edu_map = {1: 'some highschool', 2: 'graduated highschool', 3: 'some college', 4: 'graduated college'}

# Income column
income_map = {1: 'Less than $10,000', 2: '$10,000 - $14,999', 3: '$15,000 - $19,999', 
              4: '$20,000 - $24,999', 5: '$25,000 - $34,999', 6: '$35,000 - $49,999', 
              7: '$50,000 - $74,999', 8: '$75,000 or more'}


# Now apply the mappings again
diabetes_df.replace({
    'Diabetes_012': diabetes_map,
    'HighBP': bp_map,
    'HighChol': chol_map,
    'CholCheck': cholcheck_map,
    'Smoker': smoke_map,
    'Stroke': stroke_map,
    'HeartDiseaseorAttack': heart_map,
    'PhysActivity': activity_map,
    'Fruits': fruit_map,
    'Veggies': veggies_map,
    'HvyAlcoholConsump': alcohol_map,
    'AnyHealthcare': healthcare_map,
    'NoDocbcCost': dr_map,
    'GenHlth': health_map,
    'DiffWalk': walking_map,
    'Sex': sex_map,
    'Age': age_map,
    'Education': edu_map,
    'Income': income_map
}, inplace=True)

# Check the resulting DataFrame
print(diabetes_df.head())

```

```python

#column renaming 
diabetes_df.rename(columns={'Diabetes_012': 'Diabetes Status', 'CholCheck': 'Chol. Check (5 years)', 
                           'Stroke': 'Prev. Stroke', 'HeartDiseaseorAttack': 'Prev. Heart Disease or Attack', 
                           'PhysActivity': 'Physical Activity (30 Days)',  'Fruits': 'Fruit Consumption (daily)',
                           'Veggies': 'Veggie Consumption (daily)','HvyAlcoholConsump': 'Heavy Alcohol Consumption', 
                           'AnyHealthcare': 'Healthcare',  'NoDocbcCost': 'Needed but Unable to afford Dr.', 
                           'GenHlth': 'Health Rating',  'MentHlth': 'Bad Mental Health Days (30 days)', 
                           'PhysHlth': 'Bad Phys. Health Days (30 days)',  'DiffWalk': 'Walking'}, inplace=True)

diabetes_df.head()
```
